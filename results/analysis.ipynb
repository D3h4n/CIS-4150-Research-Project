```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
converters = {
    "Times": lambda x: list(map(lambda y : int(y) / 1000, x.strip().split(" ")))
}
```

```python
jsoup = pd.read_csv("jsoup.csv", converters=converters)
guava = pd.read_csv("guava.csv", converters=converters)
configs = pd.read_csv("configs.csv")
```

```python
plt.boxplot(jsoup['Times'], tick_labels=jsoup['NContainers'])
```

```python
plt.boxplot(guava['Times'], tick_labels=guava['NContainers'])
```
